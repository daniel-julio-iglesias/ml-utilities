```python
# Essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set(style='whitegrid', context='notebook')
```

```python
# Simulate energy consumption dataset
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=365, freq='D')

energy_data = pd.DataFrame({
    'date': dates,
    'electricity_usage_kWh': np.random.normal(25, 5, size=len(dates)),
    'gas_usage_m3': np.random.normal(8, 1.5, size=len(dates)),
    'water_usage_liters': np.random.normal(350, 50, size=len(dates)),
    'temperature_C': np.random.normal(15, 10, size=len(dates))
})

energy_data.head()
```

```python
# View first rows
energy_data.head()
```

```python
# Check dataset summary (statistics)
energy_data.describe()
```

```python
# Check for missing values
energy_data.isnull().sum()
```

```python
# Electricity Usage Trend
plt.figure(figsize=(12,6))
sns.lineplot(data=energy_data, x='date', y='electricity_usage_kWh')
plt.title('Daily Electricity Usage (kWh)')
plt.xlabel('Date')
plt.ylabel('Electricity (kWh)')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(energy_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
